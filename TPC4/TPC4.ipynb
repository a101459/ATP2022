```python
#Aplicação para gerir um cinema

Menu = ("Menu:\n(1) Criar Salas\n(2) Listar Salas\n(3) Verificar disponibilidade do lugar\n(4) Vender bilhete\n(5) Consultar Salas")
cinema = []
print (Menu)
option = 1

def inserirSala(cinema):
    filme = str(input("Introduza o nome do filme"))
    for i in cinema:
        nlugares, vendidos, nome = i
        if nome == filme:
            lugares = int(input("Introduza o número de lugares da sala"))
            if lugares == nlugares:
                print ("Esta sala já existe!")
                print (Menu)
                return inserirSala(cinema)
    nlugares = int(input("Introduza o número de lugares da sala"))
    vendidos = []
    cinema.append((nlugares, vendidos, filme))
    print ("A sala foi criada!")
    print ("Salas:", cinema)
    print (Menu)

def listar(cinema):
    for l in cinema:
        nlugares, vendidos, filme = l
        print ("Sala{0}".format(cinema.index(l[0:])), " ", filme[0:])
    print (Menu)

def disponivel(cinema, filme, lugar):
    for d in cinema:
        nlugares, vendidos, nome = d
        if filme != nome:
            sai = False
        else:
            if lugar in vendidos:
                sai = False
            else:
                sai = True
    return sai
    print (Menu)

def vendeBilhete(cinema, filme, lugar):
    if disponivel(cinema, filme, lugar):
        for v in cinema:
            nlugares, vendidos, nome = v
            if filme == nome:
                vendidos.append(lugar)
        print ("O lugar foi vendido!")
        print ("Salas:", cinema)
    else:
        print ("Esse lugar encontra-se indisponível.")
    print (Menu)

def listarDisponibilidades(cinema):
    for s in cinema:
        nlugares, vendidos, filme = s
        print ("Sala{0}".format(cinema.index(s[0:])), " ", filme[0:], " ", nlugares - len(vendidos))
    print (Menu)

while option != 0:
    option = int(input("Escolha uma das opções apresentadas!"))
    if option == 1:
        inserirSala(cinema)
    elif option == 2:
        listar(cinema)
    elif option == 3:
        filme = str(input("Introduza o nome do filme"))
        lugar = int(input("Introduza o número do lugar"))
        print("Lugar disponível?", disponivel(cinema, filme, lugar))
    elif option == 4:
        filme = str(input("Introduza o nome do filme"))
        lugar = int(input("Introduza o número do lugar"))
        vendeBilhete(cinema, filme, lugar)
    elif option == 5:
        listarDisponibilidades(cinema)
    elif option == 0:
        print ("Salas:", cinema)
        print ("Até à próxima!")
```
