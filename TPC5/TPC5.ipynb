```python
import csv
import pandas as pd

option = 1
Menu = ("Menu:\n(1) Ler informação do ficheiro\n(2) Tabela com informação do ficheiro\n(3) Distribuição de doença por sexo\n(4) Distribuição de doença por escalão etário\n(5) Distribuição de doença por nível de colesterol\n(6) Distribuição de doença por sexo em tabela\n(0) Encerrar programa")
info = []
print (Menu)

def lerinfo():
    file = open("myheart.csv", "r")
    informação = csv.reader(file, delimiter=",")
    line = 0
    for linha in informação:
        if line == 0:
            line += 1
        else: 
            if int(linha[5]) == 1:
                linha[5] = "Tem doença."
            else: 
                linha[5] = "Não tem doença."
            info.append((int(linha[0]),str(linha[1]),int(linha[2]),int(linha[3]),int(linha[4]),linha[5]))


def distsexo():
    file = open("myheart.csv", "r")
    informação = csv.reader(file, delimiter=",")
    line = 0
    for linha in informação:
        if line == 0:
            line += 1
        else: 
            info.append((int(linha[0]),str(linha[1]),int(linha[2]),int(linha[3]),int(linha[4]),int(linha[5])))
    countM = 0
    countF = 0
    for dados in info:
        idade,sexo,tensão,colesterol,batimento,temDoença = dados
        if temDoença == 1 and sexo == "M":
            countM += 1
        if temDoença == 1 and sexo == "F":
            countF += 1
    print ("Número de doentes do sexo masculino:", countM)
    print ("Número de doentes do sexo feminino:", countF)


def distescestários():
    file = open("myheart.csv", "r")
    informação = csv.reader(file, delimiter=",")
    line = 0
    for linha in informação:
        if line == 0:
            line += 1
        else: 
            info.append((int(linha[0]),str(linha[1]),int(linha[2]),int(linha[3]),int(linha[4]),int(linha[5])))
    escaloesEtarios = [[30,34], [35,39], [40,44], [45,49], [50,54], [55,59], [60,64], [65,69], [70,74], [75,79], [80,84], [85,89]]
    lista2 = []
    for dados in info:
        idade,sexo,tensão,colesterol,batimento,temDoença = dados
        for z in escaloesEtarios:
            if temDoença == 1 and idade >= z[0] and idade <= z[1]:
                lista2.append(z)
    for z in escaloesEtarios:
        print ("Com idades compreendidas em", z,"anos,", lista2.count(z), "pessoas têm doença.")


def distcolesterol():
    file = open("myheart.csv", "r")
    informação = csv.reader(file, delimiter=",")
    line = 0
    for linha in informação:
        if line == 0:
            line += 1
        else: 
            info.append((int(linha[0]),str(linha[1]),int(linha[2]),int(linha[3]),int(linha[4]),int(linha[5])))
    lista3 = []
    nivelcolesterol = [[0,10],[100,110], [110,120], [120,130], [130,140], [140,150], [150,160], [160,170], [170,180], [180,190], [190,200], [200,210], [210,220], [220,230],[230,240],[240,250],[250,260],[260,270],[270,280],[280,290],[290,300],[300,310],[310,320],[320,330],[330,340],[340,350],[350,360],[360,370],[370,380],[380,390],[390,400],[400,410],[410,420],[420,430],[430,440],[440,450],[450,460],[460,470],[470,480],[480,490],[490,500],[500,510],[510,520],[520,530],[530,540],[540,550],[550,560],[560,570],[570,580],[580,590],[590,600],[600,610]]
    for dados in info:
        idade, sexo, tensão, colesterol, batimento, temDoença = dados
        for p in nivelcolesterol:
            if temDoença == 1 and colesterol >= p[0] and colesterol <= p[1]:
                lista3.append(p)
    for p in nivelcolesterol:
        print ("Com nível de colesterol compreendido em", p,",", lista3.count(p), "pessoas têm doença.")


def tabela1():
    file = open("myheart.csv", "r")
    informação = csv.reader(file, delimiter=",")
    line = 0
    for linha in informação:
        if line == 0:
            line += 1
        else: 
            info.append((int(linha[0]),str(linha[1]),int(linha[2]),int(linha[3]),int(linha[4]),int(linha[5])))
    countM = 0
    countF = 0
    for dados in info:
        idade,sexo,tensão,colesterol,batimento,temDoença = dados
        if temDoença == "Tem doença." and sexo == "M":
            countM += 1
        if temDoença == "Tem doença." and sexo == "F":
            countF += 1
    listaM = [countM]
    listaF = [countF]
    tabela1 = pd.DataFrame(data =  [[listaM], 
                                    [listaF]], index = ["Homens", "Mulheres"], columns = ["Doentes"])
    print (tabela1)


while option != 0:
    option = int(input("Escolha a opção que deseja executar!"))
    if option == 1:
        lerinfo()
        print ("Os dados dos pacientes encontram-se na seguinte forma: Idade, Sexo, Tensão, Colesterol, Batimento, Tem/ Não tem doença.")
        print (info)
        print (Menu)
    elif option == 3:
        distsexo()
        print (Menu)
    elif option == 4:
        distescestários()
        print(Menu)
    elif option == 5:
        distcolesterol()
        print (Menu)
    elif option == 6:
        tabela1()
        print (Menu)
    elif option == 2:
        tab = pd.read_csv("myheart.csv")
        display(tab)
        print (Menu)
    elif option == 0: 
        print ("-------------------------------------------")
```
