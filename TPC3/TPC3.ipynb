```python
#Aplicação para manipulação de listas de inteiros

Menu = ("Menu:\n(1) Criar lista\n(2) Ler lista\n(3) Soma\n(4) Média\n(5) Maior\n(6) Menor\n(7) estaOrdenada por ordem crescente\n(8) estaOrdenada por ordem decrescente\n(9) Procurar um elemento\n(0) Sair")
print (Menu)
ListaA= []
opcao = 1
while opcao != 0:
    opcao = int(input("Escolha uma das opções apresentadas"))
    if opcao == 1:
        ListaA.clear ()
        import random
        z = random.randrange (1, 101)
        ListaA = [random.randrange(1,101) for n in range (z)]
        print (ListaA)
        print (Menu)
    elif opcao == 2:
        ListaA.clear ()
        n = int(input("Insira um valor N que corresponderá ao número de elementos da lista"))
        while (n > 0):
            x = int(input("Insira um número que deseja incluir à lista"))
            ListaA.append (x)
            n = n - 1
        print (ListaA)
        print (Menu)
    elif opcao == 3:
        soma = 0
        for i in ListaA:
            soma = soma + i
        print("A soma dos elementos da lista é",soma,".")
        print (Menu)
    elif opcao == 4:
        soma1 = 0
        for i in ListaA:
            soma1 = soma1 + i
        media = soma1 / len(ListaA)
        print("A média dos elementos da lista é",media,".")
        print (Menu)
    elif opcao == 5:
        max = ListaA[0]
        for i in ListaA[1:]:
            if i > max:
                max = i
        print ("O maior elemento da lista é o",max,".")
        print (Menu)
    elif opcao == 6:
        men = ListaA[0]
        for i in ListaA[1:]:
            if i < men:
                men = i
        print ("O menor elemento da lista é o",men,".")
        print (Menu)
    elif opcao == 7:
        ListaC= []
        for i in ListaA:
            ListaC.append(i)
        ListaC.sort()
        if ListaC == ListaA:
            print ("Sim, a lista está ordenada por ordem crescente!")
        if ListaC != ListaA:
            print ("Não, a lista não está ordenada por ordem crescente!")
        print (Menu)
    elif opcao == 8:
        ListaD= []
        for p in ListaA:
            ListaD.append(p)
        ListaD.sort(reverse = True)
        if ListaD == ListaA:
            print ("Sim, a lista está ordenada por ordem decrescente!")
        if ListaD != ListaA:
            print ("Não, a lista não está ordenada por ordem decrescente!")
        print (Menu)
    elif opcao == 9:
        procurar = int(input("Que número deseja procurar?"))
        if procurar not in ListaA:
            print ("O número não existe na lista!")
        if procurar in ListaA:
            print ("O número encontra-se na posição", ListaA.index(procurar),"da lista.")
        print (Menu)
    if opcao == 0:
        print (ListaA)
        print ("---Bye! See you soon.---")

```
